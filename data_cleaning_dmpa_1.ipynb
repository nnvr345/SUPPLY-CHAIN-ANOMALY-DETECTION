```python
import pandas as pd 
import seaborn as sns
```

```python
df= pd.read_csv("C:\\Users\\HP\\Desktop\\labs\\nnv_final_dmpa\\data_cleaning.csv")
```

```python
df.head(3)
```

```python
df.columns
```

```python
df.rename(columns={'line item insurance (usd)': 'line_item_insurance'}, inplace=True)
df.rename(columns={'unit price': 'unit_price'}, inplace=True)
df.rename(columns={'line item quantity': 'line_item_quantity'}, inplace=True)
df.rename(columns={'line item value': 'line_item_value'}, inplace=True)
df.rename(columns={'weight (kilograms)': 'weight'}, inplace=True)
df.rename(columns={'pack price': 'pack_price'}, inplace=True)
df.rename(columns={'sub classification': 'sub_classification'}, inplace=True)
df.head(3)
```

```python
df.isnull().sum()
```

```python
mean_value=df["line_item_insurance"].mean()
mean_value
```

```python
df["line_item_insurance"].fillna(mean_value,inplace=True)
```

```python
df.isnull().sum()
```

```python
df1=df.dropna()
df1.isnull().sum()
```

```python
df1.info()
```

```python
sns.boxplot(df1["line_item_insurance"])
```

```python
max_cr=df1["line_item_insurance"].quantile(0.95)
max_cr
```

```python
df1[df1["line_item_insurance"]>max_cr]
```

```python
df2=df1[df1["line_item_insurance"]<max_cr]
```

```python
sns.boxplot(df2["line_item_insurance"])
```

```python
df2.duplicated()
```

```python
df3=df2.drop_duplicates()
df3.duplicated().sum()
```

```python
attributes_to_keep = ['id','line_item_quantity','unit_price','pack_price','line_item_value','line_item_insurance','sub_classification','dosage','weight']
df3 = df3[attributes_to_keep]
```

```python
df3.to_csv("C:\\Users\\HP\\Desktop\\labs\\nnv_final_dmpa\\cleaned_data_1.csv", index=False)
```
